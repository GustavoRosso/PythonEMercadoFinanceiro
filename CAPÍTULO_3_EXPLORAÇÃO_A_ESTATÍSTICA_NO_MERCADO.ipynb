##Exemplo 3.4

```python
# Exemplo 3.4
# Análise Ibovespa
# As pequenas diferençãs nos resultados devem ser por causa do
# banco de dados das cotações, com mais ajustes.
# O autor usa a biblioteca xlrd para abrir o Excel, mas ela não funciona mais,
# então, troquei pelo Pandas.

import matplotlib.pyplot as fig
import statistics as st
import numpy as ny
import pandas as pd
import scipy.stats as sci
from scipy.stats import norm

excel = pd.read_excel('Ibv.xlsx', sheet_name = 'Ibov', header=None)
dados = list(excel[1])
lin = len(dados)


#+++++++++++++ Cálculo do retorno ++++++++++++++++++++++++++++++++
dias = ny.arange(0,lin)
dados = ny.array(dados)
retorno = (dados[1:lin] - dados[0:lin-1]) / dados[0:lin-1]
#+++++++++++++ Gráfico Ibovespa Diário +++++++++++++++++++++++++++
fig.subplot(221)
fig.plot(dias,dados, color = 'black')
fig.xlabel('Dias')
fig.title('Ibovespa Diário (2018 - Out/2019)')
#+++++++++++++ Gráfico retorno +++++++++++++++++++++++++++++++++++
fig.subplot(222)
fig.plot(dias[1:], retorno, color = 'black')
fig.xlabel('Dias')
fig.title('Ibovespa - Retorno Diário')
#+++++++++++++ Gráfico histograma e dist.normal ++++++++++++++++++
fig.subplot(223)
fig.hist(retorno, bins = 20, color = 'gray', density = True )
# Em fig.hist, normed tem que ser trocado por density.
fig.xlabel('Classes')
fig.ylabel('Frequência')
fig.title('Histograma dos retornos - Ibovespa')
xmin, xmax = fig.xlim()
media = st.mean(retorno)
desvio = st.pstdev(retorno)
eixo_x = ny.linspace(xmin, xmax,100)
eixo_y = norm.pdf(eixo_x, media, desvio)
fig.plot(eixo_x, eixo_y, color = 'black')
#+++++++++++++ Gráfico QQ-Plot +++++++++++++++++++++++++++++++++++
fig.subplot(224)
sci.probplot(retorno, dist = "norm", plot = fig)
fig.title('QQ-Plot Ibovespa')
```

```python
# Exemplo 3.9
# Regressão Linear PETR4/VALE3
# Minha solução com base nas demonstrações do autor
# As pequenas diferenças nos resultados devem ser por causa do
# banco de dados das cotações, com mais ajustes.
# O autor usa a biblioteca xlrd para abrir o Excel, mas ela não funciona mais,
# então, troquei pelo Pandas.

"""
Talvez tenha que instalar a versão antiga do xlrd com
pip install xlrd==1.2.0
Reiniciar o console, e depois sair do spyder,
quando voltar a inicializar o spyder, pode ser que funcione.
"""
import matplotlib.pyplot as fig
import xlrd
import numpy as ny
from scipy import stats
import pandas as pd

#++++++ Transfere dados da planilha para o primeiro ativo ++++++++
"""
wb1 = xlrd.open_workbook('PETR4.xlsx')
plan1 = wb1.sheet_by_name('PETR4')
lin1 = plan1.nrows
col1 = plan1.ncols
dados1 = []
"""

excel1 = pd.read_excel('PETR4.xlsx', sheet_name = 'PETR4', header=None)
dados1 = list(excel1[1])
lin1 = len(dados1)
#++++++ Transfere dados da planilha para o segundo ativo +++++++++
"""
wb2 = xlrd.open_workbook('VALE3.xlsx')
plan2 = wb2.sheet_by_name('VALE3')
lin2 = plan2.nrows
col2 = plan2.ncols
dados2 = []

for i in range(col1):
    coluna1 = plan1.col_values(i)
    dados1.append(coluna1)
    coluna2 = plan2.col_values(i)
    dados2.append(coluna2)
"""

excel2 = pd.read_excel('VALE3.xlsx', sheet_name = 'VALE3', header=None)
dados2 = list(excel2[1])
lin2 = len(dados2)
#+++++++++++++++++ Transforma lista em vetor +++++++++++++++++++
dados1 = ny.array(dados1)
dados2 = ny.array(dados2)
#+++++++++++++++++ Regressão Linear ++++++++++++++++++++++++++++
beta, beta0, r_value, p_value, std_err = stats.linregress(dados1, dados2)
print('++++++++++++++++ REGRESSÃO LINEAR +++++++++++++++')
print(' beta      beta0    r_value   p_value   std_err')
yLin = beta * dados1 + beta0
print('%10.6f %10.6f %5.2f %10.6f %10.8f' % (beta, beta0, r_value, p_value, std_err))

fig.plot(dados1, yLin, '-k', dados1, dados2, 'ok')
fig.xlabel('PETR4')
fig.ylabel('VALE3')
```
