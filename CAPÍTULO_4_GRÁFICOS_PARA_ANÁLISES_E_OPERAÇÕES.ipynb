```python
# Exemplo 4.1 - Parte 2
# Não consegui reproduzir dados da planilha Excel

from mpl_toolkits import mplot3d
import matplotlib.pyplot as fig
import numpy as ny
from scipy.interpolate import griddata as gd
import xlrd

wb = xlrd.open_workbook('Estrang.xlsx')

y = wb.sheet_by_name('Planilha1')

lin = y.nrows
col = y.ncols

dados = []

for i in range(col):
    coluna = y.col_values(i)
    dados.append(coluna)
    print(coluna)

ax = fig.axes(projection = '3d')

x = ny.arange(0, lin, 1)
z = ny.array(dados[2])
y = ny.array(dados[3])

X, Y = ny.meshgrid(x,y)
Z = gd((x,y), z, (X, Y), method = 'linear' )
ax.scatter3D(x, y, z, color = 'black');
ax.contour3D(X, Y, Z, 70, cmap = 'Greys', alpha = 1)
ax.set_xlabel('Dias', fontsize = 16)
ax.set_ylabel('Ibovespa', fontsize = 16)
ax.set_zlabel('Fluxo de Estrangeiros', fontsize = 16)
ax.set_title('Fluxo de Estrangeiros na Bovespa (2011-2012)', fontsize = 14)
```

##Exercício 4

```python
# Exercícios 4.8
# 4
# Minha solução

import matplotlib.pyplot as fig
import numpy as ny
import math

x = ny.linspace(0, 100, 50)
y = []
z = []

for i in range(len(x)):
    valorY = math.sqrt(x[i] * 2 + 1)
    y.append(valorY)
    valorZ = math.sin(x[i]) + math.sqrt(x[i]**2+1)
    z.append(valorZ)


fig.plot(x,y,'-*k', markersize = 6)
fig.plot(x,z,'--ok', markersize = 6)
fig.title('Gráfico xy', fontsize = 14)
fig.ylabel('EIXO Y', fontsize = 14)
fig.xlabel('EIXO X', fontsize = 14)
fig.grid()
```

##Exercício 5

```python
# Exercícios 4.8
# 5
# Minha solução

import matplotlib.pyplot as fig
import numpy as ny

x = ny.linspace(0, 100, 50)
y = []
z = []

for i in range(len(x)):
    valorY = 1 / ((x[i]**2)+1)
    y.append(valorY)
    valorZ = (x[i]**2) / (x[i] + 1)
    z.append(valorZ)

fig.figure()
ax1 = fig.subplot(111)
ax1.plot(x, y, '-k', linewidth = 3)
ax1.set_xlabel('X', fontsize = 14)
ax1.set_ylabel('Y', fontsize = 14)
fig.title('Gráfico xyz', fontsize = 14)
fig.grid()


ax1_segundo = ax1.twinx()

ax1_segundo.plot(x, z, '--k', linewidth = 2)
ax1_segundo.set_ylabel('Z', fontsize = 14)
```

```python
# Exercícios 4.8
# 5
# Minha solução

import matplotlib.pyplot as fig

x = ['Alta', 'Média', 'Baixa']
NTNB = [0.0398, 0.0409, 0.0401, 0.0405, 0.0406, 0.0405, 0.0405, 0.0402, 0.0401,
         0.0402, 0.0403, 0.0401, 0.0394, 0.0401, 0.0389, 0.0389, 0.0388, 0.0385,
         0.0384, 0.0374, 0.0367, 0.0368, 0.0376, 0.0370, 0.0377, 0.0373, 0.0374,
         0.0371, 0.0371, 0.0371, 0.0373, 0.0375, 0.0379, 0.0380, 0.0382, 0.0380,
         0.0376, 0.0372, 0.0369, 0.0374, 0.0370, 0.0370, 0.0372, 0.0370, 0.0377]

alta = []
media = []
baixa = []

for i in range(len(NTNB)):
    if NTNB[i] > 0.04:
        alta.append(NTNB[i])

    elif NTNB[i] <= 0.04 and NTNB[i] > 0.037:
        media.append(NTNB[i])

    else:
        baixa.append(NTNB[i])

y =[]

y.append(len(alta) /len(NTNB))
y.append(len(media) /len(NTNB))
y.append(len(baixa) /len(NTNB))

explodir = (0, 0.2, 0)
fig.pie(y, explode = explodir, labels = x, autopct='%1.1f%%')
```
