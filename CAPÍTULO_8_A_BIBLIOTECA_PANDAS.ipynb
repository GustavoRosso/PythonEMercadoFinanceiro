```python
# (b)
print('-----------------------(b)-----------------------')
b = tabela3[(tabela3['Relação crédito/depósito - EUA'] > 0.75) &
            (tabela3['Inflação anual - EUA'] < 2.5)]
print(b['Ano'])
```

```python
# (c)
print('-----------------------(c)-----------------------')
c_Brasil = tabela3['Inflação anual - Brasil'].mean()
c_EUA = tabela3['Inflação anual - EUA'].mean()

print('MÉDIA - Inflação anual - Brasil =',c_Brasil,'\nMÉDIA - Inflação anual - EUA =',c_EUA)
```

##Exercícios 15

```python
# Exercícios 8.11
# 15
# Minha solução

import pandas as pd

tabela1 = pd.DataFrame({'Ano':[1962,1963,1964,1965,1966,1967,1968,1969,1970],
                   'Preço barril petróleo (US$)':[4.01,4.01,3.96,3.85,4.01,3.69,3.56,3.56,4.01],
                   'Produção (milhões barril/dia)':[7.54,7.61,7.80,8.30,8.81,8.66,8.78,9.18,9.03]})

tabela2 = pd.DataFrame({'Ano':[1962,1963,1964,1965,1966,1967,1968,1969,1970],
                   'Dow Jones':[652,762,874,969,785,905,943,800,838],
                   'Abertura de empresas de médio porte':[56,23,48,150,17,70,170,120,150]})

# (a)
print('-----------------------(a)-----------------------')
tabela3 = pd.merge(tabela1,tabela2, on='Ano')
print(tabela3)

ax1 = tabela3.plot.line(x = 'Ano',y = ['Dow Jones','Abertura de empresas de médio porte'], color =['red','black'])
ax2 = ax1.twinx()
ax2 = tabela3.plot.line(ax = ax2, x = 'Ano',y = ['Preço barril petróleo (US$)','Produção (milhões barril/dia)'])

ax1.legend(['Dow Jones', 'Abertura de empresas de médio porte'], loc='upper left')
ax2.legend(['Preço barril petróleo (US$)', 'Produção (milhões barril/dia)'], loc='upper right')


# (b)
print('-----------------------(b)-----------------------')
b = tabela3[(tabela3['Dow Jones'] > 750) &
            (tabela3['Produção (milhões barril/dia)'] > 8.9) &
            (tabela3['Preço barril petróleo (US$)'] > 4)]

print('Abriram' + b['Abertura de empresas de médio porte'].to_string(index=False),
      'empresas nos EUA em' + b['Ano'].to_string(index=False)+'.')
```
