##12.2 ANÁLISES QUANTITATIVAS SOBRE AS INFORMAÇÕES - Explicação 1


```python
# 12.2 ANÁLISES QUANTITATIVAS SOBRE AS INFORMAÇÕES - Explicação 1

###########################################################
#                   PRECISA INSTALAR ANTES
#
#       >> pip install pytrends
###########################################################
import matplotlib.pyplot as fig
from pytrends.request import TrendReq
import matplotlib.dates as mdates
import pandas as pd

pytrends = TrendReq(hl= 'en-US', tz=360)
kw_list = ['desemprego', 'juros']

fig.style.use('ggplot')
figura = fig.figure()
ax1 = fig.subplot(111)
fig.title('GoogleTrends - Palavra chave')

pytrends.build_payload(kw_list, cat = 0, timeframe = 'today 12-m', geo = 'BR', gprop = '' )
teste = pytrends.interest_over_time()

ax1.plot(teste[kw_list[0]], '-k', teste[kw_list[1]], '--k')
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
fig.setp(ax1.get_xticklabels(),rotation = 45)
fig.legend((kw_list[0],kw_list[1]), loc = 'upper left', shadow = True)

ax1.text(x = teste.index[-1], y = teste[kw_list[0]][-1], s = kw_list[0],
         fontsize = 14, color = 'black', weight = 'bold')
ax1.text(x = teste.index[-1], y = teste[kw_list[1]][-1], s = kw_list[1],
         fontsize = 14, color = 'black', weight = 'bold')
#+++++++++++++++++ MÉDIAS QUANTITATIVAS - MÉDIA MÓVEL ++++++++++++++++
ind_med = teste.mean(axis = 1)

df = pd.DataFrame(ind_med)
df['med_mov'] = df.rolling(window = 10, min_periods = 0).mean()
ax1.plot(df['med_mov'],'-k', linewidth = 5)
```

##12.4 ESTATÍSTICA ENTRE PREÇOS E PALAVRAS-CHAVE - Explicação


```python
# 12.4 ESTATÍSTICA ENTRE PREÇOS E PALAVRAS-CHAVE - Explicação
# A um problema no servidor do Yahoo

#######################################################
#        Comparação entre preços e google trends
#######################################################
import matplotlib.pyplot as fig
from pytrends.request import TrendReq
import matplotlib.dates as mdates
import pandas_datareader as web
import datetime as dt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
#+++++++++++++++++ Conexão com o servidor ++++++++++++++++++++++++++
pytrends = TrendReq(hl = 'en-US', tz = 360)
#+++++++++++++++++ Lista de palavra-chave ++++++++++++++++++++++++++
kw_list = ['petróleo']
#++++++ Figura da frequencia das palavras-chave ++++++++++++++++++++
figura = fig.figure()
ax1 = fig.subplot(111)
fig.title('GoogleTrends - Pavavra chave')
#++++++++++++++++ Download Google trends +++++++++++++++++++++++++++
pytrends.build_payload(kw_list, cat = 0,
                       timeframe = '2018-07-01 2019-11-29',
                       geo = 'BR', gprop = '')
teste = pytrends.interest_over_time()
#+++++++++++++++++ Formatação da figura ++++++++++++++++++++++++++++
ax1.plot(teste[kw_list[0]], '--ok')
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
fig.setp(ax1.get_xticklabels(),rotation = 45)
fig.legend((kw_list[0]), loc = 'upper left', shadow = True)
ax1.text(x = teste.index[-1], y = teste[kw_list[0]][-1], s = kw_list[0],
         fontsize = 14, color = 'black', weight = 'bold')
#++++++++++++++++++++++ DataReader +++++++++++++++++++++++++++++++++
inicio = dt.datetime(2018,7,1)
fim = dt.datetime(2019,11,29)
df = web.DataReader('PETR4.SA','yahoo', inicio, fim)
ax2 = ax1.twinx()
ax2.plot(df.index, df['Close'], color = 'black')
ax2.set_ylabel('PETR4',fontsize = 18, weight = 'bold')
ax2.text(x = teste.index[73], y = 30, s = 'PETR4', fontsize = 14, weight = 'bold')
#+++++++++++ Amostragem dos pontos da ação ++++++++++++++++++++++++
med = df['Close'].resample('w').mean()
ax2.plot(med.index, med, '-k', linewidth = 2)
#+++++++++++++ Reta de Regressão Linear +++++++++++++++++++++++++++
fig.figure()
dados1 = sns.regplot(teste['petróleo'].values, med, color = 'black')
dados1.set_xlabel('Palavra "petróleo"', fontsize = 16)
dados1.set_ylabel('Preço PETR4 (R$)', fontsize = 16)
#+++++++++++++ Plotagem conjunta dos histograma das series ++++++++
dados2 = sns.jointplot(teste['petróleo'].values, med, kind = 'scatter', color = 'black')
dados2.set_axis_labels('Palavra "petróleo"','Preço PETR4 (R$)', fontsize = 16)
#+++++++++++++ Coeficiente de correlação de Pearson +++++++++++++++
dados2.annotate(stats.pearsonr)
```

```python
# 12.5 AVALIAÇÃO DE LISTAS COM DIVERSAS PALAVRAS - Explicação
# A um problema no servidor do Yahoo

#######################################################
#        Comparação entre preços e google trends
#######################################################
import matplotlib.pyplot as fig
from pytrends.request import TrendReq
import matplotlib.dates as mdates
import pandas_datareader as web
import datetime as dt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
#+++++++++++++++++ Conexão com o servidor ++++++++++++++++++++++++++
pytrends = TrendReq(hl = 'en-US', tz = 360)
#+++++++++++++++++ Lista de palavra-chave ++++++++++++++++++++++++++
kw_list = ['petróleo','dólar','crise','juros','inflação']
#++++++ Figura da frequencia das palavras-chave ++++++++++++++++++++
figura = fig.figure()
ax1 = fig.subplot(111)
fig.title('GoogleTrends - Pavavra chave')
#++++++++++++++++ Download Google trends +++++++++++++++++++++++++++
pytrends.build_payload(kw_list, cat = 0,
                       timeframe = '2018-07-01 2019-11-29',
                       geo = 'BR', gprop = '')
teste = pytrends.interest_over_time()
#+++++++++++++++++ Formatação da figura ++++++++++++++++++++++++++++
ax1.plot(teste[kw_list[0]], '--k', teste[kw_list[1]], '--k',
         teste[kw_list[2]], '--k',
         teste[kw_list[3]], '--k',
         teste[kw_list[4]], '--k')

fig.legend((kw_list[0],kw_list[1],kw_list[2],kw_list[3],kw_list[4]),
            loc = 'upper left', shadow = True)

ax1.text(x = teste.index[-1], y = teste[kw_list[0]][-1], s = kw_list[0],
         fontsize = 14, color = 'k', weight = 'bold')
ax1.text(x = teste.index[-1], y = teste[kw_list[1]][-1], s = kw_list[1],
          fontsize = 14, color = 'k', weight = 'bold')
ax1.text(x = teste.index[-1], y = teste[kw_list[2]][-1], s = kw_list[2],
         fontsize = 14, color = 'k', weight = 'bold')
ax1.text(x = teste.index[-1], y = teste[kw_list[3]][-1], s = kw_list[3],
         fontsize = 14, color = 'k', weight = 'bold')
ax1.text(x = teste.index[-1], y = teste[kw_list[4]][-1], s = kw_list[4],
         fontsize = 14, color = 'k', weight = 'bold')
#++++++++++++++++++++++ DataReader +++++++++++++++++++++++++++++++++
inicio = dt.datetime(2017,7,1)
fim = dt.datetime(2019,11,29)
df = web.DataReader('^BVSP','yahoo', inicio, fim)
ax2 = ax1.twinx()
ax2.plot(df.index, df['Close'], color = 'black')
ax2.set_ylabel('Ibovespa',fontsize = 18, weight = 'bold')
ax2.text(x = teste.index[10], y = 77000, s = 'Iovespa', fontsize = 14, weight = 'bold')
#+++++++++++ Amostragem dos pontos da ação ++++++++++++++++++++++++
ind_med = teste.mean(axis = 1)
teste['med'] = pd.DataFrame(ind_med)
med = df['Close'].resample('w').mean()
ax2.plot(med.index, med, '-k', linewidth = 2)
#+++++++++++++ Reta de Regressão Linear +++++++++++++++++++++++++++
#+++++++++++++ Plotagem conjunta dos histograma das series ++++++++
dados2 = sns.jointplot(teste['med'].values, med, kind = 'reg', color = 'black')
dados2.set_axis_labels('Média da lista de palavras','Ibovespa(pontos)', fontsize = 16)
#+++++++++++++ Coeficiente de correlação de Pearson +++++++++++++++
dados2.annotate(stats.pearsonr)
```
