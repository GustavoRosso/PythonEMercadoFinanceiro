##Exemplo 7.1

```python
# Exemplo 7.1

import datetime as dt
import matplotlib.pyplot as fig
import matplotlib.dates as mdt
import numpy as np

n = 100
y = np.random.normal(0,1,100)
hj = dt.datetime.now()
x = hj + dt.timedelta(days = n)
dias = mdt.drange(hj,x,dt.timedelta(days = 1))

fig.gca().xaxis.set_major_formatter(mdt.DateFormatter('%d-%m-%y'))
fig.gca().xaxis.set_major_locator(mdt.DayLocator(interval=5))
fig.plot(dias,y,'-k',linewidth=2)
fig.gcf().autofmt_xdate()
fig.show()
```

##Exemplo 7.2

```python
# Exemplo 7.2

import datetime as dt
import matplotlib.pyplot as fig
import matplotlib.dates as mdt
import numpy as np

data = ['01/11/2019','03/11/2019','07/11/2019']
x = [dt.datetime.strptime(i,'%d/%m/%Y').date() for i in data]
y = np.random.normal(0,1,len(x))

fig.gca().xaxis.set_major_formatter(mdt.DateFormatter('%d/%m/%Y'))
fig.gca().xaxis.set_major_locator(mdt.DayLocator())
fig.plot(x,y,'-k',linewidth=2)
fig.gcf().autofmt_xdate()
fig.show()
```

##Exemplo 7.3

```python
# Exemplo 7.3

import datetime as dt
import matplotlib.pyplot as fig
import matplotlib.dates as mdt

y = [1,5,6,8,9,10,15]
x = [dt.datetime.now() + dt.timedelta(hours=i) for i in range(len(y))]

fig.gca().xaxis.set_major_formatter(mdt.DateFormatter('%H:%M'))
fig.gca().xaxis.set_major_locator(mdt.HourLocator(interval = 2))
fig.plot(x,y,'-k',linewidth=2)
fig.gcf().autofmt_xdate()
fig.show()
```

##Exemplo 7.4

```python
# Exemplo 7.4

import datetime as dt
from datetime import datetime
import matplotlib.pyplot as fig
import matplotlib.dates as mdt
import numpy as np
import xlrd

wb = xlrd.open_workbook('Vale3.xlsx')
plan = wb.sheet_by_name('Vale')
lin = plan.nrows
col = plan.ncols

mat = np.zeros((lin,col))

for i in range(lin):
    for j in range(col):
        mat[i,j] = plan.cell(i,j).value

dataInt = mat[:,0].astype(int)
datas = [datetime.fromordinal(datetime(1900, 1,1).toordinal()+ dados_Ex - 2).strftime('%d/%m/%Y')
             for dados_Ex in dataInt]

print(datas)

ax1 = fig.subplot(111)

ax1.plot(datas[0:lin:20], mat[0:lin:20,1],'-k',linewidth = 3)

fig.gcf().autofmt_xdate()
ax1.set_xlabel('Dias',fontsize =14)
ax1.set_ylabel('VALE3',fontsize =14)
ax1.set_title('A√ß√£o VALE3 - BOVESPA')
fig.grid()
```

##Exemplo 7.5

```python
# Exemplo 7.5
# Dia do banco de dados diferente do utilizado no livro, por isso, resultado diferente

import datetime as dt
from datetime import time
import matplotlib.pyplot as fig
import matplotlib.dates as mdt
import numpy as np
import xlrd

wb = xlrd.open_workbook('BBrasil_intra.xlsx')
plan = wb.sheet_by_name('BBAS3')
lin = plan.nrows
col = plan.ncols

mat = np.zeros((lin,col))

for i in range(lin):
    for j in range(col):
        mat[i,j] = plan.cell(i,j).value

dataInt = (mat[:,0]*24*3600).astype(int)
datas = [time(dados_Ex//3600,dados_Ex%3600//60,dados_Ex%60).strftime('%H:%M:%S')
             for dados_Ex in dataInt]

print(datas)
ax1 = fig.subplot(111)
ax1.plot(datas[0:lin:20], mat[0:lin:20,1],'-k',linewidth = 3)

fig.gcf().autofmt_xdate()
ax1.set_xlabel('Horas',fontsize =14)
ax1.set_ylabel('BBSA3',fontsize =14)
ax1.set_title('A√ß√£o BBSA3 - BOVESPA')
fig.grid()
```

```python
# Exemplo 7.6

import datetime as dt
import matplotlib.pyplot as fig

f = open('datas.txt','r')
datas = []
prec = []
for linha in f:
    lista = linha.split('\n')
    x = lista[0][0:10]
    y = lista[0][12:len(lista[0])]
    datas.append(x)
    prec.append(y)

print(datas)
print(prec)
eixoY = [float(j) for j in prec]
fig.plot(datas,eixoY,'-k',linewidth=3)
fig.gcf().autofmt_xdate()
fig.grid()
```
