##Exemplo 1.2


```python
# Exemplo 1.2

import math
import statistics as st
# import xlrd             Troquei a biblioteca xlrd pela biblioteca pandas,
#                         porque xlrd não funciona mais.
import pandas as pd
import numpy as ny

excel = pd.read_excel('Petrobras.xlsx', sheet_name = 'Planilha1', header=None)

x = list(excel[0])

vetor = ny.array(x)

retorno = (vetor[1:20] - vetor[0:19]) / vetor[0:19]

media = st.mean(retorno)
desvio = st.pstdev(retorno)
otimista = media + 2 * desvio / math.sqrt(19)
pessimista = media - 2 * desvio / math.sqrt(19)

print(otimista, pessimista)
```
